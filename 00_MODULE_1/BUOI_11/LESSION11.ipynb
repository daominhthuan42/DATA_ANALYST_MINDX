```python
# Sắp xếp dataframe theo thứ tự tăng dần/giảm dần số bàn thắng
df.sort_values('Goals', ascending=False)
```

```python
# Hãy tìm ra cầu thủ nào bị nhiều thẻ vàng nhất, nhiều thẻ đỏ nhất, thi đấu nhiều nhất

# cầu thủ nào bị nhiều thẻ vàng nhất
# Option 1
df.sort_values('YellowCards', ascending=False).head(1)

#Option 2
df[df['YellowCards'] == df['YellowCards'].max()]
```

```python
# cầu thủ nào bị nhiều thẻ đỏ nhiều nhất
df[df['RedCards'] == df['RedCards'].max()]
```

```python
#  cầu thủ thi đấu nhiều nhất
df[df['Apearances'] == df['Apearances'].max()]
```

```python
# Hãy tìm ra cầu thủ thi đấu nhiều nhất của từng CLB
# Option 1
# df.sort_values('Apearances', ascending= False).drop_duplicates('Team')
# Option 2
df.loc[df.groupby('Team')['Apearances'].idxmax(), ['Team', 'Player', 'Apearances']]
```

```python
# Hãy tìm ra cầu thủ ghi nhiều bàn thắng nhất, với vị trí Defender

defenders = df.loc[df['Position'] == 'Defender']
top_defender = defenders.loc[defenders['Goals'].idxmax()]
```
